# Q4. Polynomial Plotting with Annotations:
Randomly select the coefficients of a 5th degree polynomial. Set the random seed as the last two digits of your roll number. Plot the polynomial for -5 ≤ x ≤ 5. Annotate the plot to identify the maxima and minima of the polynomial.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(13)
coefficients = np.random.randint(-10, 10, 6)  # 6 coefficients for a 5th degree polynomial
def polynomial(x, coeffs):
    return np.polyval(coeffs, x)
x_values = np.linspace(-5, 5, 400)  # 400 points between -5 and 5
y_values = polynomial(x_values, coefficients)
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Polynomial')
```

```python
# Now we find maxima and minima
maxima_indices = np.where(np.diff(np.sign(np.diff(y_values))) < 0)[0] + 1
minima_indices = np.where(np.diff(np.sign(np.diff(y_values))) > 0)[0] + 1

# Annotate maxima
for idx in maxima_indices:
    plt.annotate('Max', xy=(x_values[idx], y_values[idx]), xytext=(-20, 10), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5'), fontsize=8)

# Annotate minima
for idx in minima_indices:
    plt.annotate('Min', xy=(x_values[idx], y_values[idx]), xytext=(20, -10), textcoords='offset points',
                 arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.5'), fontsize=8)

    
plt.plot(x_values, y_values, label='Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.title('5th Degree Polynomial with Maxima and Minima Annotations')
plt.grid(True)
plt.legend()
plt.show()
```
